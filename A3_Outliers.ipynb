```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Se analiza el dataframe en búsqueda de valores nulos 
valores_nulos=dA3ta.isnull().sum()
valores_nulos
```

```python
#Se tratan los NA´s de la columna bebidas
#Lo cambiaremos por el promedio
#Hacemos una copia
dA6ta=dA5ta.copy()
dA6ta["bebidas"]=dA6ta["bebidas"].fillna(round(dA5ta["bebidas"].mean(),1))
#Comprobemos el remplazo de los NA´s por un valor específico 
valores_nulos=dA6ta.isnull().sum()
valores_nulos
```

```python
#Convertimos nuestro dataframe sin NA´s a un csv 
dA7ta.to_csv("ventas_totales_snulos1.csv")
#Se corroboran valores nulos 
valores_nulos=dA7ta.isnull().sum()
valores_nulos
data = pd.read_csv('ventas_totales_snulos1.csv')
```

# Ventas precios corrientes

```python
#Busquemos outliers en nuestras variables 
#Método desviación estándar 
#Comencemos con la columna ventas_precios_corrientes
y=data["ventas_precios_corrientes"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
#Encontramos Outliers
outliers= data[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
#No se encuentran Outliers 
```

# Ventas precios constantes 

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data["ventas_precios_constantes"]) 
plt.title("Outliers de ventas_precios_constantes")
plt.show() #dibujamos el diagrama

#Busquemos outliers en ventas_precios_constantes
y1=data["ventas_precios_constantes"]

percentile25=y1.quantile(0.25) #Q1
percentile75=y1.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)



```

```python
#Encontramos Outliers
outliers_iqr= data[(y1>Limite_Superior_iqr)|(y1<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr= data[(y1<=Limite_Superior_iqr)&(y1>=Limite_Inferior_iqr)]
data_clean_iqr
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr["ventas_precios_constantes"]) 
plt.title("Outliers de Precio Venta")
plt.show() #dibujamos el diagrama
```
